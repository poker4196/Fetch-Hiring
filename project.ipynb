```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
import itertools
%matplotlib inline

# setting up plot style 
style.use('seaborn-poster')
style.use('fivethirtyeight')
```

```python
null_col = pd.DataFrame((receipt.isnull().sum())*100/receipt.shape[0]).reset_index()
null_col.columns = ['Column Name', 'Null Values Percentage']
fig = plt.figure(figsize=(18,6))
ax = sns.pointplot(x="Column Name",y="Null Values Percentage",data=null_col,color='blue')
plt.xticks(rotation =90,fontsize =7)
ax.axhline(35, ls='--',color='red')
plt.title("Percentage of Missing values in Receipts data")
plt.ylabel("Null Values PERCENTAGE")
plt.xlabel("COLUMNS")
plt.show()
```

### Insight:
<li>Percentage of missing values of 9 features are all above 35%</li>
<li> Detected Data quality issue</li>
<li>Feature rewardsReceiptItemList must be in table Customer_Orders as the ER Diagram show to normalization table receipt</li>
